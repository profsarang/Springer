# Birch

```python
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import Birch
import pandas as pd
import seaborn as sns
```

```python
# Generating 5000 samples using make_blobs
dataset, clusters = make_blobs(n_samples = 5000, n_features = 3, cluster_std = 5, random_state = 128)
```

```python
plt.scatter(dataset[:,0], dataset[:,1],
            cmap='red',
            alpha = 0.7,
            edgecolors = 'b'
            )

plt.grid()
plt.show()
```

```python
# Creating the BIRCH clustering model
model = Birch(branching_factor = 10, n_clusters = None, threshold = 3) 
```

```python
# Fit the data (Training)
model.fit(dataset)  
```

```python

# Predict the same data
pred = model.predict(dataset)  
```

```python
# Creating a scatter plot
plt.scatter(dataset[clusters == 0, 0],
            dataset[clusters == 0, 1],
            cmap='red',
            marker='s',
            edgecolors = 'b',
            label='cluster 1')

plt.scatter(dataset[clusters == 1,0],
            dataset[clusters == 1,1],
            c ='orange',
            marker='o',
            edgecolors = 'r',
            label='cluster 2')

plt.scatter(dataset[clusters == 2,0],
            dataset[clusters == 2,1],
            c='green',
            marker='v',
            edgecolors = 'w',
            label='cluster 3')

plt.legend(loc='lower left')
plt.grid()
plt.show()
```
